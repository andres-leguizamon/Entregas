## Presentaci√≥n 
La derivada es un concepto que surge de un problema geom√©trico esencial, el problema de hallar el la tangente en un punto a una curva. Sin embargo, a partir de las elaboraciones de autores posteriores, el concepto de derivada quedo estrechamente relacionado a un problema de velocidad, de hecho , la intuici√≥n de la derivada se puede deducir a partir del concepto de velocidad en un instante. 


## Velocidad en un instante y velocidad media 
Suponga que queremos comparar como cambia la velocidad de un objeto cuando nos movemos entre dos momentos del tiempo , un momento inicial $t$ y un momento final $h$ segundos despu√©s de t, de la misma forma , podemos asumir que existe una funci√≥n $d(x)$ que nos dice a que distancia se encuentra el objeto en cada momento del tiempo. \
**Tendr√≠amos los siguientes elementos:** \
Tiempo Inicial: $t$ \
Tiempo final : $t+h$  \
Posici√≥n inicial del objeto : $f(t)$ \
Posicion final del objeto : $f(t+h)$ \

Para determinar la velocidad media entre estos momentos del tiempo se puede plantear el cociente  $\frac{{f(t+h)-f(t)}}{{h}}$ 
Cuando planteemos el limite cuando h tiende a 0 , podemos notar que los dos momentos de comparaci√≥n se van acercando hasta que se convierten en el mismo momento, por lo cual nuestra velocidad media se comezaria a acerca a la velocidad instantanea cuanto menor sea el cambio. \
Cuando estamos evaluando este limite para encontrar la velocidad instantanea, estamos encontrando la derivada de la funci√≥n.


## Ejemplo: 
Suponga que nuestra funci√≥n de velocidad es un polinomio de la forma $f(x)$= $a_nx^n+a_{n-1}x^{n-1}...a_{2}x^{2}+a_{1}x^{1}+a_{0}x^{0}$ \
Si quisieramos encontrar la derivada , o mejor dicho , la velocidad instantanea en un punto , tendriamos que plantear el limite: \

$\large\lim_{{h \to 0}} \frac{f(x+h)-f(x)}{h}\$ \

Reemplazando la forma de la funci√≥n 


$\large\lim_{{h \to 0}} \frac{a_n(x+h)^n+a_{n-1}(x+h)^{n-1}...a_{2}(x+h)^{2}+a_{1}(x+h)^{1}+a_{0}(x+h)^{0}  -a_nx^n+a_{n-1}x^{n-1}...a_{2}x^{2}+a_{1}x^{1}+a_{0}x^{0}}{h}\$ \

Al resolver algebraicamente esta suma tendremos que el resultado final es $\sum_{k=0}^{m} k a_{k}x^{k-1}$
